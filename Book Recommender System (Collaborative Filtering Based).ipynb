## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.shape # In the ratings dataset some books are not in books dataset that's why row values are decrise
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'Num_Ratings'},inplace = True)
num_rating_df['Num_Ratings'] = round(num_rating_df['Num_Ratings'],2)
num_rating_df['Num_Ratings']
```

```python
num_rating_df.sort_values(by ='Num_Ratings', ascending=False).head(20)
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'Avg_Ratings'},inplace = True)
avg_rating_df['Avg_Ratings'] = round(avg_rating_df['Avg_Ratings'],2)
avg_rating_df['Avg_Ratings']
```

```python
avg_rating_df.sort_values(by ='Avg_Ratings', ascending=False).head(20)
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df.sort_values(by ='Num_Ratings', ascending=False).head(20)
```

```python
popular_df = popular_df[popular_df['Num_Ratings']>=250].sort_values('Avg_Ratings', ascending = False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Num_Ratings','Avg_Ratings']]
```

## Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
top_rated_users = x[x].index #Boolen Indeing , facthing with index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(top_rated_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings_piovt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
final_ratings_piovt.fillna(0, inplace = True)
```

```python
final_ratings_piovt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(final_ratings_piovt)
```

```python
similarity_scores.shape
```

```python
similarity_scores[0] # first array
```

```python
sorted(list(enumerate(similarity_scores[0])),key=lambda x:x[1],reverse = True)[1:6]
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(final_ratings_piovt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse = True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == final_ratings_piovt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values)) # if you append the list they show 2-D list
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values)) # thats why we use extend
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data
```
